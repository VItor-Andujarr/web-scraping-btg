```python
# Importar Bibliotecas Necessárias
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date
```

```python
# Selecionar Opções de Abertura do Navegador
options = Options()
options.add_argument('window-size=850,850')
```

```python
# Criação de Listas para Coleta de Dados
lista_titulos = []
lista_links =[]
lista_precos = []
lista_anos = []
```

```python
pages = 3
i = 0

while i <= (pages-1):
    if i == 0:
        URL = 'https://lista.mercadolivre.com.br/veiculos/caminhoes/volkswagen/delivery-express_Desde_1_NoIndex_True'
    else:
        n = 1 + 48 * i
        URL = 'https://lista.mercadolivre.com.br/veiculos/caminhoes/volkswagen/delivery-express_Desde_{}_NoIndex_True'.format(str(n))
    print(URL)
    i+=1

    navegador = webdriver.Chrome('/Users/Vitor/Downloads/chromedriver', options=options)
    navegador.get(URL)

    sleep(5)

    page_content = navegador.page_source

    site = BeautifulSoup(page_content, 'html.parser')

    # Coleta de Dados
    produtos = site.findAll('div', attrs={'class': 'andes-card andes-card--flat andes-card--default ui-search-result shops__cardStyles ui-search-result--mot andes-card--padding-default andes-card--animated'})

    for produto in produtos:

        titulo = produto.find('h2', attrs={'class': 'ui-search-item__title ui-search-item__group__element shops__items-group-details shops__item-title'})
        lista_titulos.append(titulo.text)

        link = produto.find('a', attrs={'class': 'ui-search-link'})
        lista_links.append(link['href'])

        preco = produto.find('span', attrs={'price-tag-fraction'})
        lista_precos.append(preco.text)

        ano = produto.find('li', attrs={'ui-search-card-attributes__attribute'})
        lista_anos.append(ano.text)

        print(titulo)
        print(link['href'])
        print(preco.text)
        print(ano.text)

        print('\n')
```

```python
# Vizuaização dos dados
print(lista_titulos)
print(lista_links)
print(lista_precos)
print(lista_anos)
```

```python
# Transformação dos Dados para um DataFrame
df = pd.DataFrame({'Descrição': lista_titulos,
                   'Link': lista_links,
                   'Preço': lista_precos,
                   'Ano': lista_anos,
                   })

df.info()
```

```python
df['Preço'] = df['Preço'].replace('.', '')
df['Preço'] = df['Preço'].astype('float')
df['Ano'] = df['Ano'].astype('float')
df
```

```python
q_low = df["Preço"].quantile(0.25)
q_hi  = df["Preço"].quantile(0.75)

df = df[(df["Preço"] < q_hi) & (df["Preço"] > q_low)]
```

```python
# Exportação para um arquivo em Excel
df.to_excel(r'C:\Users\Vitor\Challenge - Dados\Delivery-Express_ML_19-10.xlsx', index = False)
```
